```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Game board
board = [" " for _ in range(9)]
buttons = []
game_over = False
output = widgets.Output()

def check_winner(b, player):
    win = [(0,1,2), (3,4,5), (6,7,8), (0,3,6), (1,4,7), (2,5,8), (0,4,8), (2,4,6)]
    return any(b[i]==b[j]==b[k]==player for i,j,k in win)

def minimax(b, is_max):
    if check_winner(b, "O"): return 1
    if check_winner(b, "X"): return -1
    if " " not in b: return 0

    if is_max:
        best = -2
        for i in range(9):
            if b[i]==" ":
                b[i]="O"
                score = minimax(b, False)
                b[i]=" "
                best = max(best, score)
        return best
    else:
        best = 2
        for i in range(9):
            if b[i]==" ":
                b[i]="X"
                score = minimax(b, True)
                b[i]=" "
                best = min(best, score)
        return best

def ai_move():
    best_score = -2
    move = 0
    for i in range(9):
        if board[i]==" ":
            board[i]="O"
            score = minimax(board, False)
            board[i]=" "
            if score > best_score:
                best_score = score
                move = i
    return move

def on_click(i):
    global game_over
    if board[i]!=" " or game_over: return

    # Player move (X)
    board[i]="X"
    buttons[i].description = "X"
    buttons[i].disabled = True

    if check_winner(board, "X"):
        with output: print("You Win!")
        game_over = True
        return
    if " " not in board:
        with output: print("Draw!")
        game_over = True
        return

    # AI move (O)
    move = ai_move()
    board[move]="O"
    buttons[move].description = "O"
    buttons[move].disabled = True

    if check_winner(board, "O"):
        with output: print("AI Wins! (Unbeatable)")
        game_over = True
    elif " " not in board:
        with output: print("Draw!")

# Create UI
for i in range(9):
    btn = widgets.Button(description=" ", layout=widgets.Layout(width='60px', height='60px'))
    btn.on_click(lambda b, i=i: on_click(i))
    buttons.append(btn)

grid = widgets.GridBox(buttons, layout=widgets.Layout(grid_template_columns="repeat(3, 60px)"))
display(grid, output)
print("You are X, AI is O. Click to play!")
```

```python
import ipywidgets as widgets
from IPython.display import display

board = [" "] * 9
btns = []
over = False
msg = widgets.Output()

def win(brd, p):
    lines = [(0,1,2),(3,4,5),(6,7,8),(0,3,6),(1,4,7),(2,5,8),(0,4,8),(2,4,6)]
    return any(brd[x]==brd[y]==brd[z]==p for x,y,z in lines)

def ai_turn():
    for i in range(9):
        if board[i]==" ":
            board[i]="O"
            if win(board,"O"): return i
            board[i]=" "
    for i in range(9):
        if board[i]==" ":
            board[i]="X"
            if win(board,"X"):
                board[i]=" "
                return i
            board[i]=" "
    if board[4]==" ": return 4
    for i in [0,2,6,8]:
        if board[i]==" ": return i
    for i in range(9):
        if board[i]==" ": return i

def click(i):
    global over
    if board[i]!=" " or over: return

    board[i]="X"
    btns[i].description="X"
    btns[i].style.button_color = '#4FC3F7'
    btns[i].disabled=True

    if win(board,"X"):
        with msg: print("You won!") ; over=True ; return
    if " " not in board:
        with msg: print("Draw!") ; over=True ; return

    m = ai_turn()
    board[m]="O"
    btns[m].description="O"
    btns[m].style.button_color = '#FF8A80'
    btns[m].disabled=True

    if win(board,"O"):
        with msg: print("AI won!") ; over=True
    elif " " not in board:
        with msg: print("Draw!") ; over=True

# UI
for i in range(9):
    b = widgets.Button(description=" ", layout=widgets.Layout(width='80px', height='80px'))
    b.style.button_color = '#E0F7FA'
    b.style.font_weight = 'bold'
    b.on_click(lambda x, i=i: click(i))
    btns.append(b)

title = widgets.HTML("<h3 style='color:#00796B'>Tic-Tac-Toe: You(X) vs AI(O)</h3>")
display(title, widgets.GridBox(btns, layout=widgets.Layout(grid_template_columns="repeat(3, 80px)", grid_gap='5px')), msg)
```
