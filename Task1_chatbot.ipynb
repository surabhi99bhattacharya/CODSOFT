```python

import random

rules = {
    "greeting": [
        ["hi", "hello", "hey"],
        ["Hey there! 😊", "Hello! How can I help?", "Hi! Nice to meet you"]
    ],
    "name": [
        ["name", "who are you", "introduce"],
        ["I'm Bolt, your rule-based buddy.", "You can call me Bolt."]
    ],
    "happy": [
        ["happy", "good", "great", "awesome", "fine", "well"],
        ["That's great to hear! 😄", "Awesome! Keep smiling!", "Glad you're feeling good!"]
    ],
    "sad": [
        ["sad", "bad", "upset", "depressed", "cry", "lonely"],
        ["I'm sorry to hear that. Want to talk about it?", "It's okay to feel down sometimes. I'm here.", "Sending you a virtual hug 🤗"]
    ],
    "angry": [
        ["angry", "mad", "frustrated", "annoyed", "irritated"],
        ["Take a deep breath. It'll be okay.", "I understand you're frustrated. Want to vent?"]
    ],
    "thanks": [
        ["thank", "thanks", "thx"],
        ["You're welcome!", "Anytime!"]
    ],
    "bye": [
        ["bye", "goodbye", "see you", "exit"],
        ["Bye! Come back soon.", "Goodbye! Take care."]
    ]
}

def get_reply(msg):
    msg = msg.lower()
    best_topic = None
    best_score = 0

    for topic, (keywords, replies) in rules.items():
        score = sum(1 for word in keywords if word in msg)
        if score > best_score:
            best_score = score
            best_topic = topic

    if best_topic:
        return random.choice(rules[best_topic][1])
    return "Hmm, I’m not trained for that yet. Try saying how you feel?"

# Chat loop
print("Bolt: Hi! I’m Bolt. Tell me how you feel, or type 'bye' to exit.")
while True:
    user = input("You: ")
    if "bye" in user.lower():
        print("Bolt:", random.choice(rules["bye"][1]))
        break
    print("Bolt:", get_reply(user))
```

```python
import random
import datetime
import re
import requests
import pytz
import wikipedia
import json
import os
from gtts import gTTS # voice

API_KEY = "d5d9ed572bf5ea18e224271d359d8b3b"
NEWS_KEY = "549b2cbbd984479ab0a14a64ff5b6e36"

# --- 1. FILE MEMORY ---
MEMORY_FILE = "buddy_memory.json"
def load_memory():
    if os.path.exists(MEMORY_FILE):
        return json.load(open(MEMORY_FILE))
    return {"name": None, "lang": "en"}
def save_memory():
    json.dump(memory, open(MEMORY_FILE, "w"))

memory = load_memory()

# --- 2. DATA ---
quotes = ["Small steps every day lead to big results.", "You are stronger than you think."]
jokes = ["Why don’t programmers like nature? Too many bugs."]
mood_rules = {
    "greeting": (["hi", "hello", "hey", "helo"], ["Hey {name}! 😊", "Hello {name}!"]),
    "sad": (["sad", "bad", "upset", "cry"], ["It's okay {name}, I'm here 🤗"]),
    "thanks": (["thank", "thanks"], ["You're welcome {name}!"])
}

def fuzzy(word, keywords):
    return any(kw in word or (len(word)==len(kw) and sum(a!=b for a,b in zip(word,kw))<=1) for kw in keywords)

# --- 3. VOICE ---
def speak(text):
    try:
        tts = gTTS(text=text, lang=memory["lang"])
        tts.save("reply.mp3")
        print("[Voice saved: reply.mp3]")
    except: pass

# --- 4. TOOLS ---
def tool_time(msg):
    ist = pytz.timezone('Asia/Kolkata')
    return "Time (IST): " + datetime.datetime.now(ist).strftime("%I:%M:%S %p")

def tool_date(msg):
    ist = pytz.timezone('Asia/Kolkata')
    return "Date: " + datetime.datetime.now(ist).strftime("%A, %d %B %Y")

def tool_calc(msg):
    expr = re.search(r'[\d\+\-\*\/\(\)\. ]+', msg)
    if expr and re.match(r'^[\d\+\-\*\/\(\)\. ]+$', expr.group().strip()):
        try: return f"Answer: {eval(expr.group())}"
        except: return "Calc error."
    return None

def tool_weather(msg):
    try:
        city = msg.split("weather in")[-1].strip().replace("?","").title() or "Kolkata"
        if city.lower() == "kolkata": city = "Calcutta"
        url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=metric"
        data = requests.get(url).json()
        return f"Weather in {city}: {data['main']['temp']}°C, {data['weather'][0]['description']}"
    except: return "Weather fetch failed."

def tool_wiki(msg): # Wikipedia
    try:
        topic = msg.split("who is")[-1].strip() or msg.split("what is")[-1].strip()
        return wikipedia.summary(topic, sentences=2)
    except: return "I couldn't find that on Wikipedia."

def tool_currency(msg): # Currency
    try:
        nums = re.findall(r'\d+', msg)
        amount = float(nums[0])
        url = "https://api.exchangerate-api.com/v4/latest/USD"
        rates = requests.get(url).json()["rates"]
        return f"{amount} USD = {amount * rates['INR']:.2f} INR"
    except: return "Try: convert 100 usd to inr"

def tool_news(msg): # News
    if NEWS_KEY == "549b2cbbd984479ab0a14a64ff5b6e36":
        return "Get free key from newsapi.org and paste above."
    try:
        url = f"https://newsapi.org/v2/top-headlines?country=in&apiKey={NEWS_KEY}"
        data = requests.get(url).json()
        headlines = [a["title"] for a in data["articles"][:3]]
        return "News: " + " | ".join(headlines)
    except: return "News fetch failed."

def tool_lang(msg): # Multi-language
    if "bangla" in msg or "bengali" in msg:
        memory["lang"] = "bn"
        save_memory()
        return "Okay, I will reply in Bengali."
    else:
        memory["lang"] = "en"
        save_memory()
        return "Okay, I will reply in English."

def tool_name_set(msg):
    memory["name"] = msg.split("my name is")[-1].strip().title()
    save_memory()
    return f"Got it, I'll call you {memory['name']}!"
def tool_name_get(msg):
    return f"Your name is {memory['name']}." if memory["name"] else "Tell me 'my name is...'"

def tool_quote(msg): return random.choice(quotes)
def tool_joke(msg): return random.choice(jokes)

tools = {
    "time": (["time"], tool_time),
    "date": (["date"], tool_date),
    "weather": (["weather"], tool_weather),
    "wiki": (["who is", "what is"], tool_wiki),
    "currency": (["convert"], tool_currency),
    "news": (["news", "headlines"], tool_news),
    "lang": (["speak in", "talk in"], tool_lang),
    "name_set": (["my name is"], tool_name_set),
    "name_get": (["what is my name"], tool_name_get),
    "motivate": (["motivate", "quote"], tool_quote),
    "joke": (["joke"], tool_joke),
}

def reply(msg):
    msg_low = msg.lower()
    calc = tool_calc(msg_low)
    if calc: return calc
    for key, (keywords, func) in tools.items():
        if any(k in msg_low for k in keywords):
            return func(msg_low)
    for topic, (keywords, replies) in mood_rules.items():
        if any(fuzzy(w, msg_low.split()) for w in keywords):
            return random.choice(replies).format(name=memory["name"] or "friend")
    return "Try: time/weather/wiki/currency/news/calculate"

print("Buddy: Hi! I'm Advance Buddy. Type 'bye' to exit.")
while True:
    user = input("You: ")
    if user.lower() in ["bye", "exit"]:
        print("Buddy: Bye!")
        break
    ans = reply(user)
    print("Buddy:", ans)
    speak(ans)
```
